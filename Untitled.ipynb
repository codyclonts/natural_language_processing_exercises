```python
import unicodedata
import re
import json

import nltk
from nltk.tokenize.toktok import ToktokTokenizer
from nltk.corpus import stopwords

import pandas as pd

import acquire
```

```python
#1 basic clean function

def basic_clean(df):
    df = df.lower()
    df = unicodedata.normalize('NFKD', df)\
    .encode('ascii', 'ignore')\
    .decode('utf-8', 'ignore')
    df = df.replace
    df = re.sub(r"[^a-z0-9'\s]", '', df)
    return df
```

```python
# 2 tokenize function
tokenizer = nltk.tokenize.ToktokTokenizer()
```

```python
def tokenize(df):
    return(tokenizer.tokenize(df, return_str=True)[0:500])
```

```python
# 3 
def stem(df):
    ps = nltk.porter.PorterStemmer()
    for word in df:
        print('stem:', ps.stem(word))
```

```python
# 4 
def lemmatize(df):
    wnl = nltk.stem.WordNetLemmatizer()
    for word in df:
        print('stem:', wnl.stem(word))
```
